# ai_heartbeat dataset 

```python
import os
import shutil

# Define source and destination directories
source_dir = '../data/source/ai_heartbeat/'
healthy_dir = '../data/dataset/healthy/'
unhealthy_dir = '../data/dataset/unhealthy/'

# Create destination directories if they don't exist
os.makedirs(healthy_dir, exist_ok=True)
os.makedirs(unhealthy_dir, exist_ok=True)

# List of directories to process
healthy_dirs = ['Atraining_normal', 'Btraining_normal', 'Btraining_normal_noisy']
unhealthy_dirs = [
    'Atraining_murmur'
    'Btraining_extrasystole',
    'Btraining_murmur',
    'Btraining_murmur_noisy',    
]

# Helper function to copy .wav files from a source directory to a destination directory
def copy_wav_files(source_subdir, destination_dir):
    source_path = os.path.join(source_dir, source_subdir)
    if os.path.isdir(source_path):
        for root, _, files in os.walk(source_path):
            for file in files:
                if file.endswith('.wav'):
                    src_file = os.path.join(root, file)
                    dest_file = os.path.join(destination_dir, file)
                    shutil.copy2(src_file, dest_file)
                    print(f"Copied: {src_file} to {dest_file}")
    else:
        print(f"Directory not found: {source_path}")

# Copy healthy .wav files
for directory in healthy_dirs:
    copy_wav_files(directory, healthy_dir)

# Copy unhealthy .wav files
for directory in unhealthy_dirs:
    copy_wav_files(directory, unhealthy_dir)

print("File operations complete.")
```

# heart_sound dataset

```python
import os
import shutil

# Define source directories for heart_sound
heart_sound_train_healthy = '../data/source/heart_sound/train/healthy'
heart_sound_val_healthy = '../data/source/heart_sound/val/healthy'
heart_sound_train_unhealthy = '../data/source/heart_sound/train/unhealthy'
heart_sound_val_unhealthy = '../data/source/heart_sound/val/unhealthy'

# Define destination directories
destination_healthy = '../data/dataset/healthy'
destination_unhealthy = '../data/dataset/unhealthy'

# Function to copy .wav files from a source directory to a destination directory
def copy_files(source_dir, destination_dir, postfix):
    if os.path.isdir(source_dir):
        for root, _, files in os.walk(source_dir):
            for file in files:
                if file.endswith('.wav'):
                    src_file = os.path.join(root, file)
                    filename, ext = os.path.splitext(file)
                    dest_file = os.path.join(destination_dir, f"{filename}{postfix}{ext}")
                    shutil.copy2(src_file, dest_file)
                    print(f"Copied: {src_file} to {dest_file}")
    else:
        print(f"Directory not found: {source_dir}")

# Ensure destination directories exist
os.makedirs(destination_healthy, exist_ok=True)
os.makedirs(destination_unhealthy, exist_ok=True)

# Copy healthy files from train and val to the destination healthy directory
copy_files(heart_sound_train_healthy, destination_healthy, "-train")
copy_files(heart_sound_val_healthy, destination_healthy, "-val")

# Copy unhealthy files from train and val to the destination unhealthy directory
copy_files(heart_sound_train_unhealthy, destination_unhealthy, "-train")
copy_files(heart_sound_val_unhealthy, destination_unhealthy, "-val")

print("All files copied successfully.")
```

# Heartbeat sounds dataset

```python
import os
import shutil
import pandas as pd

data_dir = '../data/source/heartbeat_sounds/'

# Define paths to the datasets and CSV files
set_a_csv_path = os.path.join(data_dir, 'set_a.csv')
set_b_csv_path = os.path.join(data_dir, 'set_b.csv')
output_dir = os.path.join('..', 'data', 'dataset')

# Create the output directory
os.makedirs(output_dir, exist_ok=True)

def process_dataset(source_dir, csv_path):
    df = pd.read_csv(csv_path)
    
    for index, row in df.iterrows():
        label = row['label']
        fname = row['fname']
        
        if label and type(label) is not float:
            
            
            fname = fname.replace("Btraining_", "")
             
            # Create a directory for the class if it doesn't exist
            class_dir = os.path.join(output_dir, str(label))
            os.makedirs(class_dir, exist_ok=True)
            
            # Construct the full path to the source audio file
            file_path = os.path.join(source_dir, fname)
            
            # fix error in one data labeling case
            folder, file = fname.split("/")
            file = file.replace("_", "__", 1)
            f_name_double = folder + "/" + file
            file_path_double = os.path.join(source_dir, f_name_double)
            
            # Copy the audio file to the corresponding class folder
            if os.path.exists(file_path):
                shutil.copy(file_path, class_dir)
            elif os.path.exists(file_path_double):
                # fix error in data labeling
                shutil.copy(file_path_double, class_dir)
            else:
                print(f"File not found: {file_path}")

# Process both datasets
process_dataset(data_dir, set_a_csv_path)
process_dataset(data_dir, set_b_csv_path)
```

```python
# extrahls, extrastole and murmur into unhealthy, normal into healthy

import os
import shutil

# Define directories
normal_dir = '../data/dataset/normal'
healthy_dir = '../data/dataset/healthy'
murmur_dir = '../data/dataset/murmur'
extrastole_dir = '../data/dataset/extrastole'
extrahls_dir = '../data/dataset/extrahls'
unhealthy_dir = '../data/dataset/unhealthy'

# Rename files from normal to healthy
if os.path.exists(normal_dir):
    # Create healthy directory if it doesn't exist
    os.makedirs(healthy_dir, exist_ok=True)

    # Move and rename files
    for file_name in os.listdir(normal_dir):
        source_file = os.path.join(normal_dir, file_name)
        destination_file = os.path.join(healthy_dir, file_name)
        if os.path.isfile(source_file):
            shutil.move(source_file, destination_file)
            print(f"Moved: {source_file} to {destination_file}")
    shutil.rmtree(normal_dir)

# Move files from murmurs, extrastole, and extrahls to unhealthy
for dir_path in [murmur_dir, extrastole_dir, extrahls_dir]:
    if os.path.exists(dir_path):
        # Create unhealthy directory if it doesn't exist
        os.makedirs(unhealthy_dir, exist_ok=True)

        for file_name in os.listdir(dir_path):
            source_file = os.path.join(dir_path, file_name)
            destination_file = os.path.join(unhealthy_dir, file_name)
            if os.path.isfile(source_file):
                shutil.move(source_file, destination_file)
                print(f"Moved: {source_file} to {destination_file}")

        # Remove the original directory after moving files
        shutil.rmtree(dir_path)
        print(f"Removed directory: {dir_path}")

print("File operations complete.")
```

```python
import os
import shutil

# Define source and destination directories
source_dir = '../data/source/ESC-50-master/audio'
destination_dir = '../data/dataset/artifact'

# Check if the source directory exists
if not os.path.exists(source_dir):
    print(f"Source directory does not exist: {source_dir}")
else:
    # List all files in the source directory
    files = os.listdir(source_dir)

    # Copy each file to the destination directory
    for file_name in files:
        # Construct full file path
        source_file = os.path.join(source_dir, file_name)
        destination_file = os.path.join(destination_dir, file_name)

        # Check if it's a file (not a directory)
        if os.path.isfile(source_file):
            shutil.copy(source_file, destination_file)
            print(f"Copied: {source_file} to {destination_file}")
        else:
            print(f"Skipped (not a file): {source_file}")

print("File transfer complete.")
```

# Split datasets

```python
import os
import random
import shutil

def split_dataset(input_dir, output_dir, val_ratio=0.1, test_ratio=0.1, seed=42):
    random.seed(seed)
    for class_name in os.listdir(input_dir):
        class_path = os.path.join(input_dir, class_name)
        if os.path.isdir(class_path):
            files = os.listdir(class_path)
            random.shuffle(files)

            total_files = len(files)
            val_count = int(total_files * val_ratio)
            test_count = int(total_files * test_ratio)
            
            splits = {
                'train': files[val_count + test_count:],
                'val': files[:val_count],
                'test': files[val_count:val_count + test_count]
            }

            for split, file_list in splits.items():
                split_dir = os.path.join(output_dir, split, class_name)
                os.makedirs(split_dir, exist_ok=True)
                for file in file_list:
                    shutil.copy(os.path.join(class_path, file), os.path.join(split_dir, file))

split_dataset('../data/dataset', '../data/splitted_dataset')
```
